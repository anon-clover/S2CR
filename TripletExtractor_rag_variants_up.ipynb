```python
from config import *
from Reader import aReader
import json

aLLM = aReader(
    # temperature=0.,
)
# OPENAI_API_BASE_URL
```

```python
"""处理数据"""
#  检索信息的三元组提取
TE_prompt_doc = """ prompt
"""
# answer的三元组提取
TE_prompt_ans = """prompt
"""

with open('../dataset/Biography/input.txt','r', encoding='utf-8') as f:
    text = f.read()
    datas = text.split("\n\n\n")

    datas = [data.split('\n')[:2] for data in datas]
    datas = [[rows.split(': ')[1] for rows in data] for data in datas]
  
base_ans = json.load(open("../results/Biography/AG.json", "r",encoding='utf-8'))
en_ans = json.load(open("../results/Biography/AG.json","r",encoding='utf-8'))
print(len(base_ans))
```

```python
doc_prompts = [TE_prompt_doc.format(query='', documents=data[1]) for data in datas]

base_prompts = [TE_prompt_ans.format(query=datas[i][0], answer=base_ans[i]) for i in range(len(datas))]
en_prompts = [TE_prompt_ans.format(query=datas[i][0], answer=en_ans[i]) for i in range(len(datas))]
```

```python
retr_triple = aLLM.generate_list_responses(doc_prompts, progress_bar=True)
ans_triple_base = aLLM.generate_list_responses(base_prompts, progress_bar=True)
ans_triple_en = aLLM.generate_list_responses(en_prompts, progress_bar=True)
```

```python
import json

json.dump(retr_triple, open("../results/Biography/Retr_Triple.json",'w',encoding='utf-8'), ensure_ascii=False)
json.dump(ans_triple_base, open("../results/Biography/LLaMa_70b/wo_retr/Ans_Triple_base.json",'w', encoding='utf-8'), ensure_ascii=False)
json.dump(ans_triple_en, open("../results/Biography/LLaMa_70b/w_retr/Ans_Triple_EN.json",'w', encoding='utf-8'), ensure_ascii=False)
```
