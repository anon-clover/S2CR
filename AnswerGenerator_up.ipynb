```python
"""处理数据"""

AnswerGenPrompt_base = """prompt\n"
                      Question: {query}\n"
                        Retrieval Information: {documents}\n"
"""
AnswerGenPrompt_en = """prompt\n"
                      Question: {query}\n"
                       Retrieval Information: {documents}\n"
"""

with open('../dataset/ALCE-ASQA/input.txt','r', encoding='utf-8') as f:
    text = f.read()
    datas = text.split("\n\n\n")
    datas = [data.split('\n')[:2] for data in datas]
    datas = [[rows.split(': ')[1] for rows in data] for data in datas]
print(len(datas))
```

```python
base_prompts = [AnswerGenPrompt_base.format(query=data[0], documents=data[1]) for data in datas]
en_prompts = [AnswerGenPrompt_en.format(query=data[0], documents=data[1]) for data in datas]
```

```python
import nest_asyncio
nest_asyncio.apply()

base_AG = aLLM.generate_list_responses(base_prompts)
en_AG = aLLM.generate_list_responses(en_prompts)
```
